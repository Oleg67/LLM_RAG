### Load authors

```python
with open('./data/data_rag/conference/conference.json', 'r') as f: 
    meta_data = json.load(f) 
```

```python
from prompt_utills import *
```

```python
written_author = PromptTemplate.from_template(
            """
            You are an expert in analyzing scientific papers.
            You have received an author's name of some  scientific paper:
            ###
                {author}
            ------------- \n
            Your task is to think of other ways this name could be written.
            For example, a different order of the first and last name, some letters changed, some abbreviations, etc.
            Make a list of possible spellings of this name.
            
            The output format is a list in valid JSON format.
            Example: ["John Smith", "Smith John", "john smith", "smith john"]
            """
        )
```

```python
authors = written_author | llm | JsonOutputParser()
```

```python
d_a = dict()
for d  in meta_data:
    for a in  d.get('authors', dict()).keys():
        answer = authors.invoke({'author': a})
        for k in answer:
            d_a[k] = a
    print(d.get('authors', dict()).keys())
    print(len(d_a),"\n\n")
```
